## Defining Type

```python
type_1= len(set(Sub1))
type_2= len(set(Sub2))
type_3= len(set(Sub3))

types=[type_1,type_2,type_3]
print(types)
```

## Lexical Diversity

```python
def lexDiv(types,tokens):
    diversity= types/tokens
    return diversity

print("lexical diversity subcorpa 1", lexDiv(type_1,length_sub_1))
print("lexical diversity sbcorpa 2", lexDiv(type_2,length_sub_2))
print("lexical diversity sub3", lexDiv(type_3,length_sub_3))
```

## Frequency Distributions 

```python
# Tokenize the text (Sub1)
text = Sub1  # Replace with your actual text
tokens1 = word_tokenize(text)
freq_dist = FreqDist(tokens1)

print("First Subcorpus top 10 words",freq_dist.most_common(13))
print("---------------------------------------")
text2 = Sub2  # Replace with your actual text
tokens2 = word_tokenize(text2)
freq_dist2 = FreqDist(tokens2)
print("Second Subcorpus top 10 words ",freq_dist2.most_common(13))
print("---------------------------------------")
text3 = Sub3  # Replace with your actual text
tokens3 = word_tokenize(text3)
freq_dist3 = FreqDist(tokens3)

print("Third Subcorpus top 10 words",freq_dist3.most_common(13))
```
